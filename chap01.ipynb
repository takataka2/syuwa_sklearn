```python
# 機械学習の機械学習が得意なこと

最近よく「機械学習」という言葉を耳にします。そもそも機械学習とはどのような技術なのでしょうか。研究者によって定義は様々ですが、
人間の学習能力をコンピュータで実現しようとする技術と思ってもらえば良いです。大量のデータからパターンや法則を見つけ出し、
有益な情報が得られるプログラムを作ることを目的としています。

通常のプログラムと機械学習を使ったプログラムの違いを確認するために、一つ身近な機械学習の応用例を見て見ましょう。

将棋や囲碁のプログラムです。チェスの世界チャンピオンのがリル・カスパロフが
IBMで開発されたチェス専用コンピュータ「ディープブルー」に破れたのは1997年のことです。それ以来将棋、囲碁でも対戦プログラムの研究が進められていたのですが、
なかなかプロレベルの強さまで到達することができませんでした。チェスに比べて将棋や囲碁のプログラムの研究者・開発者が少なかったこともあるのですが、
チェスより想定できる局面が膨大なためコンピュータでしらみつぶしに局面を調べる手法は通用しませんでした。

将棋や囲碁のプログラムのポイントはいくつかありますが、重要なのは局面の評価、つまりある局面が有利か不利かを正しく判断することです。
局面の有利、不利が分かれば、次の手を選ぶときに有利な局面を選び続ければ自然と勝利できるはずです。

従来のプログラムでは局面評価を次のように処理していました。
まず評価基準となる項目をたくさん用意しておきます。手番を握っている、自玉が堅い、駒得している、詰みがあるなどです。
それぞれの項目の重要度を数値にして おきます。次に局面を評価基準に照らし合わせ、重要度を考慮して 評価値を計算します。
これを敵味方それぞれで計算して比較すれば、 有利か不利かがわかります。あとは、自分に有利な局面になる手を指 していきます。

各項目の重要度をどのように設定するかが開発者の腕の見せ所であり、長年の経験をもとに調整していました。

２００５年、将棋のプログラムにブレイクスルーが起こります。機械学習を取り入れた「Bonanza(ボナンザ)」の登場です。
Bonanzaはプロが指した将棋やネットで公開されている将棋の棋譜をプログラムが読み込んで学習し、学習した情報を元に
項目の重要度を調整する仕組みを備えていました。どういうことかというと、ある局面をボナンザが形勢判断するときに、
プロの棋士が指した手をボナンザが選ぶように項目の重要度を調整するのです。何万、何十万という局面でプロが指した手を
選ぶように訓練されたプログラムはプロ並みの実力を持つという理屈です。

その後Bonannzaのプログラムは公開されたことで、機械学習を取り込んだプログラムが多数出現して将棋ソフトの実力は
急激に上昇しました。そして2017年に将棋の名人を倒し、完全にソフトが人間を上回りました。

以上の例からもわかるように機械学習が得意とするのは次のような問題です。

- 手作業による大量の調整や、ルールの長いリストが必要な場合。
- 伝統的な手法では複雑すぎて手に負えない問題。
- 状況が絶えず変動する。機械学習ならば２４時間体制で新しいデータに対応が可能。
```

```python
## 機械学習とは何か

機械学習の他の応用例として迷惑メールを削除してくれるスパム（迷惑メール）フィルターをみてみましょう。
スパムフィルターは受信したメールを正常なメールと迷惑メールに分類します。
まず従来のスパムフィルターの処理をまとめると次のようになります。

- スパムを分析します。本文、タイトル、送信元などに使われている単語から、スパムの特徴となるパターンを見つけ出します。
- スパムの特徴をルールにまとめます。例えば文面に「当選しました」という文字が使われている、送信元が特定のメールアドレスならばスパムとする、など。
- 上記のルールに該当するメールにスパムのフラグ（しるし）をつけます。

従来のプログラムの問題点は、スパムの特徴を読み取り、ルールにまとめる手間が膨大な点です。しかも迷惑メールは常に同じ内容とは限りません。
使われる単語が変わったり、新しくスパムを送りつける業者が現れればそのたびにルールを更新していく必要があります。

機械学習を使ったスパムフィルターの処理をまとめると次のようになります。
- 正常とスパムに分類されたメールのテキストデータをプログラムが大量に読み込ます。
- スパムに頻繁に登場する単語や特徴を読み取って、できるだけ高確率でスパムと判定できるように判定プログラムを調整します。
- 調整した判定プログラムで該当するメールにスパムのフラグをつけます。

機械学習のアプローチをとったスパムフィルターは、人間が特に手を出さなくても勝手に「当選しました」という単語が入ったメールや
怪しいメールアドレスから送られてきたメールをスパムと判定します。いちいち人間がスパムメールを研究したり、ルールを書き換える必要がなくなるので負担が減ります。
しかも最新のスパムメールのデータを与え続ければスパム判定のルールも常に更新されていきます。スパムメールとのイタチごっこに人間が悩むことはなくなります。

このスパムフィルターは機械学習を取り入れたプログラムの典型的な例です。システムが学習のために使うデータを訓練セット（training set）と呼びます。
そして個々のデータ例を訓練インスタンス、あるいは標本、サンプルと呼びます。


```
